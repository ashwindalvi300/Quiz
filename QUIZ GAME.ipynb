```python
import random

class QuizGame:
    def __init__(self, questions):
        self.questions = questions
        self.score = 0

    def display_question(self, question):
        print(question["question"])
        for i, option in enumerate(question["options"], start=1):
            print(f"{i}. {option}")
        user_answer = input("Your answer (enter the number): ")
        return int(user_answer)

    def run_quiz(self):
        print("Welcome to the Quiz Game!")
        for question in self.questions:
            user_answer = self.display_question(question)
            if user_answer == question["correct_answer"]:
                print("Correct!")
                self.score += 1
            else:
                correct_option = question["options"][question["correct_answer"] - 1]
                print(f"Sorry, the correct answer was {question['correct_answer']}: {correct_option}")
            print("---------------")

        print(f"\nQuiz completed! Your final score is: {self.score}/{len(self.questions)}")

# Define datasets for easy and hard questions
easy_questions = [
    {
        "question": "What is the capital of France?",
        "options": ["Berlin", "Madrid", "Paris", "Rome"],
        "correct_answer": 3,
        "difficulty": "easy"
    },
    {
        "question": "Which planet is known as the Red Planet?",
        "options": ["Mars", "Venus", "Jupiter", "Saturn"],
        "correct_answer": 1,
        "difficulty": "easy"
    }
]

hard_questions = [
    {
        "question": "Who wrote 'One Hundred Years of Solitude'?",
        "options": ["Gabriel Garcia Marquez", "Leo Tolstoy", "Charles Dickens", "Jane Austen"],
        "correct_answer": 1,
        "difficulty": "hard"
    },
    {
        "question": "In which year did World War II end?",
        "options": ["1943", "1945", "1950", "1939"],
        "correct_answer": 2,
        "difficulty": "hard"
    }
]

# Prompt the user to choose a difficulty level
user_choice = input("Choose a difficulty level (easy/hard): ").lower()

if user_choice == "easy":
    selected_questions = easy_questions
elif user_choice == "hard":
    selected_questions = hard_questions
else:
    print("Invalid choice. Please choose either 'easy' or 'hard'.")
    exit()

# Shuffle the selected questions for variety
random.shuffle(selected_questions)

# Create a QuizGame instance and run the quiz
quiz_game = QuizGame(selected_questions)
quiz_game.run_quiz()
```
